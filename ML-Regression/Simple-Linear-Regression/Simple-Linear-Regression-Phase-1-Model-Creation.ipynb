```python
# Importing the pandas library for data manipulation and analysis
import pandas as pd
```

```python
# Reading CSV file using pandas library and storing it in the variable salary_data
dataset = pd.read_csv("Salary_Data.csv")
```

```python
# printing dataset values
dataset
```

```python
# Extracting 'YearsExperience' column from salary_data as independent
# Using double square brackets to ensure the result is a 2D format
independent = dataset[["YearsExperience"]]
# printing independent value
independent
```

```python
# Extracting the 'Salary' column from salary_data as the target variable (y)
dependent = dataset[["Salary"]]
# printing dependent values
dependent
```

```python
# Importing the train_test_split function from sklearn.model_selection package to split data into training and testing sets
from sklearn.model_selection import train_test_split
# Splitting the data into training (70%) and testing (30%) subsets
# random_state ensures maintaining consistent order of inputs
x_train, x_test, y_train, y_test = train_test_split(independent, dependent, test_size=0.30, random_state=0)
```

```python
# Importing the LinearRegression function from sklearn.linear_model package to build a linear regression model
from sklearn.linear_model import LinearRegression
# Creating an instance of the LinearRegression class (regressor) to build the model
regressor = LinearRegression()
# Training the linear regression model using the training data
regressor.fit(x_train, y_train)
```

```python
# This represents the change in the target variable(expected to increase salary for every unit)
weight = regressor.coef_
weight
```

```python
# Retrieving the intercept value from the trained model, when YearsExperience is zero(assuming starting value).
bias = regressor.intercept_
bias
```

```python
# Predicting the target values (Salary) for the "test dataset" using the trained model
y_pred = regressor.predict(x_test)
y_pred
```

```python
# Importing the r2_score function from sklearn.metrics to evaluate the model
from sklearn.metrics import r2_score
# Calculating the R-squared score to measure how the model's predictions match the actual test data
r2_score = r2_score(y_test, y_pred)
r2_score
```

```python
# Importing the pickle library to save Python objects to a file
import pickle
```

```python
# Assigning file name in a variable
filename = "Linear_Model.sav"
# Using pickle to save the trained model to the specified file.
pickle.dump(regressor, open(filename, "wb"))
```

```python
# Loading model from saved file
model_load = pickle.load(open(filename, "rb"))
# Using model to predict 
model_load.predict([[14]])
```
